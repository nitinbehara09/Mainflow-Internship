```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hello = pd.read_csv('/content/USvideos.csv')
```

```python
hello.head()
```

```python
hello.shape
```

```python
hello = hello.drop_duplicates()
hello.shape
```

```python
hello.describe()
```

```python
hello.info()
```

```python
columns_to_remove = ['thumbnail_link','description']
hello = hello.drop(columns=columns_to_remove)
hello.info()
```

```python
from datetime import datetime
```

```python
import datetime
```

```python
hello["trending_date"] = hello["trending_date"].apply(lambda x : datetime.datetime.strptime(x,'%y.%d.%m'))
hello.head(3)
```

```python
hello['publish_time']=pd.to_datetime(hello['publish_time'])
hello.head(2)
```

```python
hello['publish_month'] = hello['publish_time'].dt.month
hello['publish_day'] = hello['publish_time'].dt.day
hello['publish_hour'] = hello['publish_time'].dt.hour
hello.head(2)
```

```python
print(sorted(hello["category_id"].unique()))
[1,2,10,15,17,19,20,222,23,24,25,26,27,28,29,30,43]
```

```python
import numpy as np
hello['category_name'] = np.nan
hello.loc[(hello["category_id"] == 1), "category_name"] = 'Film and Animation'
hello.loc[(hello["category_id"] == 2), "category_name"] = 'Autos and Vehicles'
hello.loc[(hello["category_id"] == 10), "category_name"] = 'Music'
hello.loc[(hello["category_id"] == 15), "category_name"] = 'Pets and Animals'
hello.loc[(hello["category_id"] == 17), "category_name"] = 'Sports'
hello.loc[(hello["category_id"] == 19), "category_name"] = 'Travel and Events'
hello.loc[(hello["category_id"] == 20), "category_name"] = 'Gaming'
hello.loc[(hello["category_id"] == 22), "category_name"] = 'People and Blogs'
hello.loc[(hello["category_id"] == 23), "category_name"] = 'Comedy'
hello.loc[(hello["category_id"] == 24), "category_name"] = 'Entertainment'
hello.loc[(hello["category_id"] == 25), "category_name"] = 'News and Politics'
hello.loc[(hello["category_id"] == 26), "category_name"] = 'How to and Style'
hello.loc[(hello["category_id"] == 27), "category_name"] = 'Education'
hello.loc[(hello["category_id"] == 28), "category_name"] = 'Science and Technology'
hello.loc[(hello["category_id"] == 29), "category_name"] = 'Non Profits and Activism'
hello.loc[(hello["category_id"] == 30), "category_name"] = 'Movies'
hello.loc[(hello["category_id"] == 43), "category_name"] = 'Shows'

hello.head()
```

```python
hello['year']=hello['publish_time'].dt.year
yearly_counts = hello.groupby('year')['video_id'].count()

yearly_counts.plot(kind='bar' , xlabel='Year', ylabel='Total Publish Count', title='Total Publih Video Per year')

plt.show
```

```python
yearly_views=hello.groupby('year')['views'].sum()

yearly_views.plot(kind='bar' , xlabel='Year', ylabel='Total views', title='Total views Per year' )
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='category_name', data=hello,order=hello['category_name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Video Count by Category')
plt.show
```

```python
category_views = hello.groupby('category_name')['views'].sum().reset_index()
top_categories = category_views.sort_values(by='views', ascending=False).head(5)
plt.bar(top_categories['category_name'], top_categories['views'])
plt.xlabel('Category Name', fontsize=12)
plt.ylabel('Total Views', fontsize=12)
plt.title('Top 5 Categories', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
videos_per_hour = hello['publish_hour'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.barplot(x=videos_per_hour.index, y=videos_per_hour.values, palette='rocket')
plt.title('Number of Videos Published per Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.show()
```

```python
hello['publish_time'] = pd.to_datetime(hello['publish_time'])
hello['publish_date'] = hello ['publish_time'].dt.date
video_count_by_date = hello.groupby('publish_date').size()
plt.figure(figsize=(12, 6))
sns.lineplot(data=video_count_by_date)
plt.title("Videos Published Over Time")
plt.xlabel('Publish Date')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=hello,x='views',y='likes')
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
plt.figure(figsize = (14,8))
plt.subplots_adjust(wspace=0.2, hspace=0.4, top = 0.9)
plt.subplot(2,2,1)
g = sns.countplot(x='comments_disabled', data=hello)
g.set_title("Comments Disabled", fontsize=16)
plt.subplot(2,2,2)
g1=sns.countplot(x='ratings_disabled', data=hello)
g1.set_title("Rating Disabled", fontsize=16)
plt.subplot(2,2,3)
g2=sns.countplot(x='video_error_or_removed', data=hello)
g2.set_title("Video Error or Removed", fontsize=16)
plt.show()
```
